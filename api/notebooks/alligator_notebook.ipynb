```python
import requests
import os

BASE_ENDPOINT = os.environ["ALLIGATOR_NOTEBOOK_ENDPOINT"]
TOKEN = os.environ["ALLIGATOR_NOTEBOOK_TOKEN"]  

headers = {'accept': 'application/json'}
params = {'token': TOKEN}
```

```python
file_name = "film" # name of the csv file
files = {'file': (f'{file_name}.csv', open(f'{file_name}.csv', 'rb'), 'text/csv')}

dataset_name = "test" # give a name to the dataset
ENDPOINT = f"{BASE_ENDPOINT}/dataset/{dataset_name}/table"
response = requests.post(ENDPOINT, files=files, headers=headers, params=params)

print(response.status_code)
print(response.json())  # If the response is in JSON format
```

```python
ENDPOINT = f"{BASE_ENDPOINT}/dataset/{dataset_name}/table/{file_name}"
response = requests.get(ENDPOINT, headers=headers, params=params)


print(response.status_code)
print(response.json())  # If the response is in JSON format
result = response.json()
```

```python
import pandas as pd

def get_dict_cea_data(cea):
    entities = {}
    for item in cea:
        id_row, id_col = (item["idRow"], item["idColumn"])
        if len(item["entity"]) > 0:
            entities[f"{id_row} {id_col}"] = item["entity"][0]
    return entities   


headers = []
rows = []
i = 0
entities = get_dict_cea_data(result["semanticAnnotations"]["cea"])
for row in result["rows"]:
    cells = []
    id_row = row["idRow"]
    for id_col, cell in enumerate(row["data"]):
        cells.append(cell)
        entity = entities.get(f"{id_row} {id_col}")
        if entity is not None:
            cells.append(entity["id"])
            cells.append(entity["name"])
            cells.append(entity["score"])
        i+=1
    rows.append(cells)    

#pd.DataFrame([item["data"] for item in ["rows"]])
```
